#### The Ted Website is not allowing to scrape all the pages but it can go to some limit and then stop. so we tried repeating the scraper for multiple times through our manual page number given and it scrapes the data through it 

```python
def scraper(url):
    all_data = pd.DataFrame()  # Create an empty DataFrame to store all data
    page = 151
    
    while True:
        # Create the URL for the current page
        current_url = f"{url}?page={page}"
        
        # Send a GET request to the URL
        response = requests.get(current_url)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all the div elements with class 'col-xs-6 title' and 'col-xs-2 event'
        title_elements = soup.find_all('div', class_='col-xs-6 title')
        event_elements = soup.find_all('div', class_='col-xs-2 event')

        # Initialize empty lists for 'Title' and 'Event'
        title = []
        event = []

        # Extract the values from these div elements and append to the respective lists
        for title_element, event_element in zip(title_elements, event_elements):
            title_span = title_element.find('span', {'class': 'l3'})
            event_span = event_element.find('a')
            if title_span and event_span:
                title_text = title_span.get_text(strip=True)
                event_text = event_span.get_text(strip=True)
                title.append(title_text)
                event.append(event_text)

        # Create a DataFrame with 'Title' and 'Event' columns for the current page
        data = {'Title': title, 'Event': event}
        df = pd.DataFrame(data)
        
        # Concatenate the data to the main DataFrame
        all_data = pd.concat([all_data, df], ignore_index=True)
        
        print(f"Scraped data from Page {page}")
        
        # Check if the "Next" option is available
        next_button = soup.find('a', class_='pagination__next pagination__flipper pagination__link')
        if not next_button:
            break  # Exit the loop if there is no "Next" button
        
        page += 1  # Move to the next page

    return all_data

# Example usage
ted_url = "https://www.ted.com/talks/quick-list"  # Replace with your TED Talks URL
result_df = scraper(ted_url)

# Print the resulting DataFrame
print(result_df)
```

```python
#check if the generated content available in the corpus

for i in range(10):
    generated_title = model.make_short_sentence(max_chars=1000)
    if new_content.isin([generated_title]).sum() == 1:
        print(True)
    else:
        print(generated_title)
```
